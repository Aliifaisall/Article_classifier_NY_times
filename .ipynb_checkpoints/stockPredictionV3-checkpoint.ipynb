```python
#importing json to be able to read data given
import json
#image processing libraries 
import os
import numpy as np
#pip install Pillow
from PIL import Image
# TfidfVectorizer so that i can convert raw tet into a matrix of TF-IDF features
#Term Frequency-Inverse Document Frequency
#refrecnce : https://medium.com/@cmukesh8688/tf-idf-vectorizer-scikit-learn-dbc0244a911a
#pip install scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
```

```python
# defining a function for preprocessing text
# Refrence 
def preprocess_text(text):
    #https://vaclavkosar.com/ml/Tokenization-in-Machine-Learning-Explained
    #tokenization
    tokens = word_tokenize(text.lower())
    #Remove punctuation
    tokens = [word for word in tokens if word not in string.punctuation] 
    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    # Join the tokens back into a single string
    preprocessed_text = ' '.join(tokens)
    
    return preprocessed_text
```

```python
#Read the json file
with open(r'C:\git\Rihal\N24News\news\nytimes_dev.json') as f:
    data = json.load(f)
```

```python
# Extract relevant images from JSON
image_folder = r'C:\git\Rihal\N24News\imgs'
image_id_to_file = {os.path.splitext(file)[0]: os.path.join(image_folder, file) 
                    for file in os.listdir(image_folder)}
image_features = []
```

```python
sections = []
preprocessed_texts = []
```

```python
#construct a loop that iterate through the data , to extract relevant information
#to feed my machine learning model section, headline, article, and image ID
for item in data: 
    section = item.get('section', '') 
    headline = item.get('headline','')
    article = item.get('article', '')

    #preprocess text
    preprocessed_article = preprocess_text(article)
    preprocessed_texts.append(preprocessed_article)
    sections.append(section)

    #Extract image features for the model
    image_id = item.get('image_id', '')
    image_file = image_id_to_file.get(image_id,'')
    if image_file and os.path.exists(image_file):
        image = Image.open(image_file)
        # feature etraction
        image_feature = np.random.rand(1000)
        image_features.append(image_feature)
    else:
        #if not found use the placeholder features straight away
        image_feature.aapend(np.zeros(1000))
```

```python
# this line converts the image features into NmuPy array
image_features = np.array(image_features)
```

```python
# Combine text and image features
X_text = preprocessed_texts
X_combined = np.concatenate((image_features, TfidfVectorizer(max_features=1000).fit_transform(X_text).toarray()), axis=1)
y = sections
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)
```

```python
# Train a RandomForestClassifier with cross-validation
clf = RandomForestClassifier(n_estimators=100, random_state=42)
cv_scores = cross_val_score(clf, X_train, y_train, cv=5)
print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", np.mean(cv_scores))
```

```python
# Fit the classifier on the training data
clf.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred = clf.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Test set accuracy:", accuracy)
```
